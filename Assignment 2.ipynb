Task 3: Create a function to calculate the distance between two coordinates using the haversine formula. Write the following formula where the input parameters are a pair of coordinates as two lists

```python
from math import sin, cos, sqrt, atan2, radians
# Radius of the earth(km)
R = 6371

# coordinates of St Andrews and my home town

lat1 = radians(56.33871)
lon1 = radians(-2.79902)
lat2 = radians(53.70326)
lon2 = radians(-2.32382)

# calculating the difference in coordinates

dlon = lon2 - lon1
dlat = lat2 - lat1

# equation to calculate distance

a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
c = 2 * atan2(sqrt(a), sqrt(1 - a))

distance = R * c
print("Result: ", distance)
```

Q5. • Calculate a new column named "pop_M" (population in millions), by transforming the "pop" (population) column.
• Remove the original "pop" column.
• Choose/subset a city that starts with the same letter as your first name (for example,
"Mexico City" if your first name is Michael).
• Subset the five biggest (i.e., largest population sizes) cities from the country where your
selected city is.
• Print the result.
• Make sure you write the scripts in different code cells where you describe the process.

```python
#loading the data 

import pandas as pd
worldcities_df = pd.read_csv("Assignment 2 Data/world_cities.csv")
worldcities_df
```

```python
# making "pop_M" column 

worldcities_df['pop_M'] = worldcities_df.loc[:, 'pop']
```

```python
# turning the data into millions

def to_million(x):
    return x/1000000
worldcities_df.pop_M = worldcities_df.pop_M.apply(to_million)
worldcities_df
```

```python
# delete original pop column 

worldcities_million_df = worldcities_df.drop('pop', axis=1)
worldcities_million_df
```

```python
# subset Austin texas

just_Austin = worldcities_million_df[worldcities_million_df["city"] =="Austin"]
print(just_Austin)
```

```python
# subset the data to just USA cities

cities_USA = worldcities_million_df.query('country=="USA"')
print(cities_USA)
```

```python
#subset 5 largest cities in the USA

largestcities_USA = cities_USA.nlargest(5, ['pop_M'])
print(largestcities_USA)
```
